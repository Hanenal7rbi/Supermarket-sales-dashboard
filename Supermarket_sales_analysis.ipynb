```python
monthly_sales= df.groupby(df['Date'].dt.to_period('M')).agg({'Total':'sum'})
print(monthly_sales)
```

#Add a column


```python
df['Day of Week'] = df['Date'].dt.day_name()
df.info
```

```python
df['Time'] = pd.to_datetime(df['Time'], 
format='mixed')
df['Hour of Day'] = df['Time'].dt.hour
```

```python
df.info
```

```python
df.groupby('Payment')['Total'].sum().plot(kind='bar')
plt.title('Total Sales by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Total Sales')
plt.show()
```

```python
df.groupby(['Branch', 'City'])['Total'].sum().unstack().plot(kind='bar')
plt.title('Total Sales by Branch and City')
plt.xlabel('Branch')
plt.ylabel('Total Sales')
plt.show()
```

```python
df.groupby('Product line')['Total'].sum().plot(kind='bar')
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.show()
```

```python
df.groupby('Customer type')['Total'].sum().plot(kind='bar')
plt.title('Total Sales by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Total Sales')
plt.show()
```

```python
monthly_sales.plot(kind='line')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Hour of Day'].value_counts().plot(kind='bar')
plt.title('Most Common Purchase Times of the Day')
plt.xlabel('Hour of Day')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Day of Week'].value_counts().plot(kind='bar')
plt.title('Most Common Purchase Days of the Week')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.show()
```
